```python
file_path = []
for pictures in os.listdir(folder_path):
    path_pictures = os.path.join(folder_path,pictures)
    file_path.append(path_pictures)
file_path
```

```python
vector = []
for images in file_path:
    embedded = cv2.imread(images)
    vector.append(embedded)    
vector
```

```python
depth = len(vector)
rows = len(vector[0])/depth
cols = len(vector[0][0])/depth
print("Rows: ",rows)
print("cols:",cols)
print("depth: ",depth)
```
